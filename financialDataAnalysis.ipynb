***Question 4 (15 pts)***

4.a) Return the top 3 companies in each STATE based on the average "Long-Term Debt – Total" (5 pts)
* You can use a user-defined function following the lecture notes
* You will need to first aggregate at the company level to account for the fact that a company appears multiple times due to the column YEAR.
* The unique identifier of company in the dataset is GVKEY. You can also use TICKER or Company Name

```python
# write code here
def top(df, column, n=3):
    return df.sort_values(by=column)[-n:]

FirmAndExecutivesFinancialDataAggStateCorp = FirmAndExecutivesFinancialData.groupby([FirmAndExecutivesFinancialData['STATE'], FirmAndExecutivesFinancialData['Company Name']]).mean()
FirmAndExecutivesFinancialDataAggStateCorp = FirmAndExecutivesFinancialDataAggStateCorp.reset_index()

FirmAndExecutivesFinancialDataAggState = FirmAndExecutivesFinancialDataAggStateCorp.groupby(FirmAndExecutivesFinancialDataAggStateCorp['STATE'])
FirmAndExecutivesFinancialDataAggState.apply(top, column='Long-Term Debt – Total', n=3)[['Long-Term Debt – Total']]
```

4.b) Return the top 3 companies in each STATE led by MALE CEOs based on the average "Long-Term Debt – Total" (5 pts)
* You can use a user-defined function following the lecture notes
* GENDER contains the gender of CEOs
* The unique identifier of company in the dataset is GVKEY. You can also use TICKER or Company Name

```python
# write code here
def top(df, column, n=3):
    return df.sort_values(by=column)[-n:]

FirmAndExecutivesFinancialDataGENDER = FirmAndExecutivesFinancialData[FirmAndExecutivesFinancialData["GENDER"] == 'MALE']

FirmAndExecutivesFinancialDataGENDERAggStateCorp = FirmAndExecutivesFinancialDataGENDERAggStateCorp.groupby([FirmAndExecutivesFinancialDataGENDERAggStateCorp['STATE'], FirmAndExecutivesFinancialDataGENDER['Company Name']]).mean()
FirmAndExecutivesFinancialDataGENDERAggStateCorp = FirmAndExecutivesFinancialDataGENDERAggStateCorp.reset_index()

FirmAndExecutivesFinancialDataGENDERAggState = FirmAndExecutivesFinancialDataGENDERAggStateCorp.groupby(FirmAndExecutivesFinancialDataGENDERAggStateCorp['STATE'])
FirmAndExecutivesFinancialDataGENDERAggState.apply(top, column='Long-Term Debt – Total', n=3)[['Company Name','Long-Term Debt – Total']]
```

```python
# write code here
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy

```

```python
y, X = patsy.dmatrices('Q("Long-Term Debt – Total") ~ ZIP + GENDER + SIC + CITY + ADDRESS', FirmAndExecutivesFinancialData)
model_1 = sm.OLS(y, X)
results_model_1 = model_1.fit()
print(results_model_1.summary())

```
